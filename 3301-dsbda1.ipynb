```python
wine_df = pd.read_csv("winequalityN.csv")
```

```python
wine_df.head()
```

```python
#checking for missing values
wine_df.isnull().sum()
```

```python
# Replace missing values with the median value of the column
# Use one-hot encoding to convert the "type" column to numerical values
wine_df = pd.get_dummies(wine_df, columns=['type'])
```

```python
# Drop the original "type" column
wine_df = wine_df.drop(['type_white'], axis=1)
```

```python
# Convert the remaining columns to float
wine_df = wine_df.astype(float)
```

```python
wine_df.fillna(wine_df.median())
```

```python
wine_df.shape
```

```python
# Convert the "quality" variable to a categorical variable
wine_df['quality'] = pd.Categorical(wine_df['quality'])
wine_df['quality'] = wine_df['quality'].cat.codes
```

```python
wine_df["quality"]
```

```python
wine_df.dtypes
```

```python
#creating a dictionary to map categorical values to numeric values
quality_dict = {"3":1,"4":2,"5":3,"6":4,"7":5,"8":6}
wine_df["quality"] = wine_df["quality"].map(quality_dict)
```
