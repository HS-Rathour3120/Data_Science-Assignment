```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import the Iris dataset
df = pd.read_csv('Iris-2.csv')
```

```python
df
```

```python
# Preprocessing: Drop any rows with missing values
df.dropna(inplace=True)
```

```python
# Data formatting
df['Species'] = df['Species'].astype('category')

# Convert all numeric columns to float data type
numeric_cols = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
df[numeric_cols] = df[numeric_cols].astype(float)

# Print out the column names and data types
print("Features and their types:")
print(df.dtypes)
```

```python
# Creating a histogram for each feature
df.hist(bins=10, figsize=(10,8))
plt.show()
```

```python
# Creating a boxplot for each feature
fig, axs = plt.subplots(ncols=4, figsize=(15,6))
axs[0].boxplot(df['SepalLengthCm'])
axs[0].set_title('Sepal Length')
axs[1].boxplot(df['SepalWidthCm'])
axs[1].set_title('Sepal Width')
axs[2].boxplot(df['PetalLengthCm'])
axs[2].set_title('Petal Length')
axs[3].boxplot(df['PetalLengthCm'])
axs[3].set_title('Petal Width')
plt.show()
```

```python
sns.boxplot(data=df,x='SepalLengthCm')
```

```python
df['Species'] = df['Species'].replace({'Iris-setosa':1,'Iris-verginica':2,'Iris-versicolor':3})
from scipy import stats

z_scores = stats.zscore(df)
outliers = df[(z_scores > 3).any(axis=1)]

print("Outliers:")
print(outliers)
```
