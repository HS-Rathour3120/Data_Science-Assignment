```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the Titanic dataset into a Pandas DataFrame
titanic_df = pd.read_csv('train.csv')
```

```python
# Perform data preprocessing and cleaning
titanic_df.drop(['Cabin', 'Ticket', 'Name'], axis=1, inplace=True)
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)
```

```python
# Data formatting
titanic_df['Sex'] = titanic_df['Sex'].map({'female': 1, 'male': 0})
titanic_df['Embarked'] = titanic_df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
# Converting to suitable type
titanic_df['Age'] = titanic_df['Age'].astype(int)
```

```python
# Visualize the data using Seaborn
# sns.countplot(x='Survived', data=titanic_df)
sns.catplot(x='Survived', y='Age', data=titanic_df, kind='box', hue='Sex')
# sns.catplot(x='Pclass', y='Survived', data=titanic_df, kind='bar', hue='Sex')
```

```python
# Plot a histogram of the ticket prices
plt.hist(titanic_df['Fare'], bins=20)
plt.xlabel('Ticket Price')
plt.ylabel('Frequency')
plt.title('Distribution of Ticket Prices')
plt.show()
```
