```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the Titanic dataset into a Pandas DataFrame
titanic_df = pd.read_csv('train.csv')
```

```python
titanic_df.isnull().sum()
```

```python
# Perform data preprocessing and cleaning
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
titanic_df.drop(['Cabin', 'Ticket', 'Name'], axis=1, inplace=True)
titanic_df = titanic_df.dropna(subset=["Embarked"])
```

```python
# Data formatting
titanic_df['Sex'] = titanic_df['Sex'].map({'female': 1, 'male': 0})
titanic_df['Embarked'] = titanic_df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
titanic_df['Age'] = titanic_df['Age'].astype(int)
```

```python
# Plot a box plot of age with respect to each gender and survival status
sns.boxplot(x='Sex', y='Age', hue='Survived', data=titanic_df)
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Distribution of Age by Gender and Survival Status')
plt.show()
```

```python
# Calculate median age of survival for male passengers
male_survivors = titanic_df[(titanic_df["Sex"] == 0) & (titanic_df["Survived"] == 1)]
male_median_age = male_survivors["Age"].median()
print("Median age of survival for male passengers:", male_median_age)
```

```python
# Calculate median age of survival for female passengers
female_survivors = titanic_df[(titanic_df["Sex"] == 1) & (titanic_df["Survived"] == 1)]
female_median_age = female_survivors["Age"].median()
print("Median age of survival for female passengers:", female_median_age)
```
